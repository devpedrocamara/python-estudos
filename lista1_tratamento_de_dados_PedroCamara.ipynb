## (f) Identificar e listar as posições (células) contendo valores que não pertencem ao domínio esperado
Explicação: aqui precisamos definir o domínio esperado por coluna. Vamos:
- inspecionar tipos inferidos por pandas
- para colunas numéricas verificar strings ou valores não-convertíveis
- para colunas de data verificar parse failures
- para colunas categóricas verificar valores fora do conjunto provável (ex.: campos que deveriam ser 'int', 'float', 'str', 'datetime')


```python
# 1. Visualize os tipos de dados atuais
print("Tipos de dados originais:")
print(df.dtypes)
print("-" * 30)

def identificar_valores_tipo_incorreto(df):
    """
    Identifica as células que não se enquadram no tipo de dado 'inferido'
    ou esperado para a coluna.
    """
    posicoes_incorretas = []

    # Iterar sobre todas as colunas
    for coluna in df.columns:
        # Pega o tipo de dado original da coluna para inferir o que deve ser checado
        tipo_original = df[coluna].dtype

        # --- Lógica de Verificação por Tipo de Dado ---

        # 1. Colunas que deveriam ser Numéricas (int, float)
        if tipo_original in (np.dtype('int64'), np.dtype('float64')) or 'int' in str(tipo_original) or 'float' in str(tipo_original):
            # Tenta forçar a conversão para numérico. Valores não numéricos viram NaN.
            coluna_coercida = pd.to_numeric(df[coluna], errors='coerce')

            # Identifica onde a conversão falhou (valores que viraram NaN mas NÃO eram NaN originalmente)
            # Para evitar contar valores ausentes (NaN) já existentes como "incorretos"
            mascara_incorreta = coluna_coercida.isna() & df[coluna].notna()

            # Adiciona as posições à lista de resultados
            if mascara_incorreta.any():
                indices = df.index[mascara_incorreta].tolist()
                for i in indices:
                    valor = df.loc[i, coluna]
                    posicoes_incorretas.append({
                        'Linha (Índice)': i,
                        'Coluna': coluna,
                        'Valor Encontrado': valor,
                        'Tipo Esperado (Inferido)': 'Numérico',
                        'Observacao': f"Não é um número válido. Tipo original: {tipo_original}"
                    })

        # 2. Colunas que deveriam ser Datas (DateTime)
        elif 'datetime' in str(tipo_original):
            # Tenta forçar a conversão para datetime. Valores não formatáveis viram NaT.
            coluna_coercida = pd.to_datetime(df[coluna], errors='coerce', dayfirst=True)

            # Identifica onde a conversão falhou (valores que viraram NaT mas NÃO eram NaT/NaN originalmente)
            mascara_incorreta = columna_coercida.isna() & df[coluna].notna()

            # Adiciona as posições à lista de resultados
            if mascara_incorreta.any():
                indices = df.index[mascara_incorreta].tolist()
                for i in indices:
                    valor = df.loc[i, coluna]
                    posicoes_incorretas.append({
                        'Linha (Índice)': i,
                        'Coluna': coluna,
                        'Valor Encontrado': valor,
                        'Tipo Esperado (Inferido)': 'Data/Tempo',
                        'Observacao': f"Não está no formato de data/hora esperado. Tipo original: {tipo_original}"
                    })

        # 3. Colunas Categóricas/Texto (object)
        # Para colunas de texto (object), a verificação de "tipo incorreto" é mais subjetiva.
        # Geralmente, só se verifica se há algum dado numérico ou data "solto" no meio.
        # Aqui, vamos focar em verificar se há apenas números ou apenas datas que não foram detectados corretamente
        elif tipo_original == np.dtype('object'):
             # Tenta coerção para numérico
            coluna_numerica = pd.to_numeric(df[coluna], errors='coerce')
            if coluna_numerica.notna().all() and df[coluna].notna().any():
                 # Se todos os valores NÃO NaN puderem ser convertidos para numérico, a coluna é "inconsistente"
                 # pois está como 'object' mas deveria ser numérica.
                 pass # Poderia adicionar um alerta aqui, mas não é um erro de CÉLULA

            # Tenta coerção para data
            coluna_data = pd.to_datetime(df[coluna], errors='coerce', dayfirst=True)
            if coluna_data.notna().all() and df[coluna].notna().any():
                # Se todos os valores NÃO NaN puderem ser convertidos para data, a coluna é "inconsistente"
                # pois está como 'object' mas deveria ser data.
                pass # Poderia adicionar um alerta aqui, mas não é um erro de CÉLULA

        # IMPORTANTE: A verificação de 'object' é limitada.
        # O principal objetivo é encontrar strings em colunas numéricas ou datas.

    return posicoes_incorretas

# --- Execução ---

resultados = identificar_valores_tipo_incorreto(df)

if resultados:
    print("\n[ RESULTADOS ]")
    print("Foram encontradas células com valores fora do domínio (tipo de dado) esperado:")

    # Converte a lista de dicionários para um DataFrame para melhor visualização
    df_resultados = pd.DataFrame(resultados)

    # Exibe o DataFrame de resultados
    with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', 1000):
        print(df_resultados)
else:
    print("\n[ RESULTADOS ]")
    print("Nenhuma célula foi identificada com tipo de dado inconsistente após a coerção para os tipos esperados (numérico/data).")
```

l) Eliminar as linhas cujo valor da coluna “text” contenham “trava-zaps”.

```python
import re

# Função para identificar mensagens suspeitas de 'trava-zaps'
def is_travazap(text):
    if pd.isna(text):
        return False

    # Texto muito longo
    if len(text) > 1000:  # você pode ajustar o limite
        return True

    # Muitas palavras com apenas caracteres especiais
    if len(re.findall(r'[\W_]', text)) / max(len(text), 1) > 0.3:  # >30% caracteres não alfanuméricos
        return True

    # Sequências repetitivas de símbolos (ex: emojis ou caracteres especiais)
    if re.search(r'(.)\1{10,}', text):  # mesma caractere repetido 10+ vezes
        return True

    return False

# Aplicar filtro e remover linhas suspeitas
df = df[~df['text_content_anonymous'].apply(is_travazap)]

# Conferir resultado
df[['text_content_anonymous']].head(10)
```

m) Identificar inconsistências entre os atributos (features).

```python
# Lista para armazenar inconsistências
inconsistencias = []

# 2.1 - Checar caracteres x text
mask = df['caracteres'] != df['text_content_anonymous'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
for idx in df[mask].index:
    inconsistencias.append((idx, 'caracteres vs text'))

# 2.2 - Checar words x text
mask = df['words'] != df['text_content_anonymous'].apply(lambda x: len(str(x).split()) if pd.notna(x) else 0)
for idx in df[mask].index:
    inconsistencias.append((idx, 'words vs text'))

# 2.3 - Checar has_media vs media_url
mask = (df['has_media'] == False) & (df['media_url'].notna())
for idx in df[mask].index:
    inconsistencias.append((idx, 'has_media vs media_url'))

# 2.4 - Checar viral vs sharings
mask = (df['sharings'] == 1) & (df['viral'] == 1)
for idx in df[mask].index:
    inconsistencias.append((idx, 'viral vs sharings'))

# 2.5 - Checar sentiment vs score_sentiment
mask = ((df['sentiment'] == 1) & (df['score_sentiment'] <= 0)) | \
       ((df['sentiment'] == -1) & (df['score_sentiment'] >= 0))
for idx in df[mask].index:
    inconsistencias.append((idx, 'sentiment vs score_sentiment'))

# Mostrar resultado
print(inconsistencias)
```
